## Relevant Columns
### Founders:
#### Industry
#### Verticals
#### Geography
#### Investment Stage
#### Business Revenue Model
### Investors:
#### Industry
#### Vertical
#### Preferred Geography
#### Preferred Investment Stage
#### Preferred Investment Type

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# Founders
founder_vectors = []
for index, row in founders.iterrows():
  founder_text = " ".join([row['Industry'], row['Verticals'], row['Geography'], row['Investment Stage'], row['Business Revenue Model']])
  founder_vec = model.encode(founder_text)
  founder_vectors.append(founder_vec)

# Investors
investor_vectors = [] 
for index, row in investors.iterrows():
  investor_text = " ".join([row['Industry'], row['Vertical'], row['Preferred Geography (from Fund from Fund Master)'], row['Preferred Investment Stage'], row['Preferred Investment Type']])
  investor_vec = model.encode(investor_text)
  investor_vectors.append(investor_vec)
  
# Use cosine similarity on these vectors for recommendations
from sklearn.metrics.pairwise import cosine_similarity

# Convert lists to numpy arrays
founder_vectors = np.array(founder_vectors)
investor_vectors = np.array(investor_vectors)

# Calculate cosine similarity
similarity_matrix = cosine_similarity(founder_vectors, investor_vectors)

# Get top 5 recommendations for each founder
top_5_recommendations = np.argsort(similarity_matrix, axis=1)[:, -5:]

# Print recommendations
for i, founder in enumerate(founders['Name']):
    print(f"Top 5 investor recommendations for {founder}:")
    for j in top_5_recommendations[i]:
        print(investors['Investor Name'][j])
    print("\n")
```

```python
# write founder investor matching as a json file
founder_investor_matching = {}
for i, founder in enumerate(founders['Name']):
    founder_investor_matching[founder] = [investors['Investor Name'][j] for j in top_5_recommendations[i]]

# Write the dictionary to a JSON file
import json
with open('founder_investor_matching.json', 'w') as f:
    json.dump(founder_investor_matching, f)
```

```python
# Get top 5 recommendations for each Investor
top_5_recommendations_investors = np.argsort(similarity_matrix, axis=0)[-5:, :]

# Print recommendations
for i, investor in enumerate(investors['Investor Name']):
    print(f"Top 5 founder recommendations for {investor}:")
    for j in top_5_recommendations_investors[:, i]:
        print(founders['Name'][j])
    print("\n")
```

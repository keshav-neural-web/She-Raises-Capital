# Feature Engineering

```python
# Define the mapping between investment stages in founders and investors data
investment_mapping = {
    'Pre-SEED': ['SEED', 'Early Stage'],
    'SEED': ['Early Stage'],
    'Series D': ['Late Stage'],
    # Add more mappings based on domain knowledge
}

# Iterate over each row in the founders' data
for index, row in founders.iterrows():
    investment_stage = row['Investment Stage']
    new_investment_stages = [investment_stage]  # Start with the original investment stage
    
    # Check if the investment stage has a mapping in the investment_mapping dictionary
    if investment_stage in investment_mapping:
        new_investment_stages.extend(investment_mapping[investment_stage])
    
    # Update the "Investment Stage" column with the new comma-separated values
    founders.at[index, 'Investment Stage'] = ', '.join(new_investment_stages)
```

## Weighted Similarity

```python
# Founders
founder_vectors = []
for index, row in founders.iterrows():
  founder_text = " ".join([row['Industry'], row['Verticals'], row['Geography'], row["Target Market (Locations)"], row['Investment Stage'], row['Business Revenue Model'], row['Market size']])
  founder_vec = model.encode(founder_text)
  founder_vectors.append(founder_vec)

# Investors
investor_vectors = [] 
for index, row in investor_fund_master.iterrows():
  try:
    investor_text = " ".join([row['Industry'], row['Vertical'], row['Preferred Geography (from Fund from Fund Master)'], row['Preferred Investment Stage'], row['Preferred Investment Type'], row["Preferred TAM"]])
    investor_vec = model.encode(investor_text)
    investor_vectors.append(investor_vec)
  except:
     pass
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Define a function to calculate cosine similarity with prioritized features
def calculate_similarity(founder_vector, investor_vector, priority_order):
    # Calculate cosine similarity
    similarity = cosine_similarity(founder_vector.reshape(1, -1), investor_vector.reshape(1, -1))[0][0]
    
    # Apply priority weights based on feature order
    weighted_similarity = similarity
    for feature in priority_order:
        if feature in founder_vector and feature in investor_vector:
            weighted_similarity *= 2  # Increase similarity if the feature matches
    
    return weighted_similarity

# Create a list to store recommendations for each founder
all_recommendations = []

# Loop through each founder and calculate recommendations
for founder_vector in founder_vectors:
    recommendations = []
    for i, investor_vector in enumerate(investor_vectors):
        # Define the priority order for features (industry and vertical should have higher priority)
        priority_order = ['Industry', 'Vertical', 'Preferred Geography', 'Preferred Investment Stage', 'Preferred TAM']
        
        # Calculate weighted similarity
        similarity = calculate_similarity(founder_vector, investor_vector, priority_order)
        
        recommendations.append((i, similarity))

    # Sort recommendations by similarity score (highest first)
    recommendations.sort(key=lambda x: x[1], reverse=True)
    
    # Get the top 5 recommendations
    top_5_recommendations = recommendations[:5]
    
    # Append the recommendations for this founder to the list
    all_recommendations.append(top_5_recommendations)

# Print recommendations for each founder
for founder_index, recommendations in enumerate(all_recommendations):
    founder_name = founders.iloc[founder_index]['Name']
    print(f"Recommendations for {founder_name}:")
    
    for investor_index, similarity_score in recommendations:
        investor_name = investor_fund_master.iloc[investor_index]['Fund from Fund Master']
        print(f"Investor: {investor_name}, Similarity Score: {similarity_score:.4f}")
    
    print("\n")
```
